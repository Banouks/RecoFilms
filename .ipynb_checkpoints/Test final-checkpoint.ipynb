```python
def recherche_casting2(des_mots, un_df) :
    
    try : 
        recherche = des_mots.lower().split()
    
        for mot in recherche : 
            un_df = un_df.loc[un_df['primaryName'].str.lower().\
                              str.contains(mot)][['tconst', 'primaryName', 'popularity']]
    
        return list(un_df.sort_values(by= 'popularity', ascending= False).head(6)['tconst'])
    except :
        return print('Aucun film trouvé !')

test = recherche_casting2('bra pit', df_films)
display(test)
```

```python
def recherche_casting(un_id, un_df):
    récuperation = un_df[un_df['tconst'] == un_id]['primaryName']\
        .apply(lambda x : x.split(sep=','))
    liste = récuperation.iloc[0]
    if len(liste) > 4 :
        liste = liste[:4]
    return liste

def rech_film_par_cast(une_liste, un_index, un_df) :
    un_df = un_df.loc[un_df['primaryName'].str.contains(une_liste[un_index])]
    un_df = un_df.sort_values(by= 'popularity', ascending= False)['tconst'].head(6)
    films = un_df.to_list()
    return films

acteurs = recherche_casting('tt2713180', df_films)
print(acteurs)
test = rech_film_par_cast(acteurs, 1, df_films)
test
```
